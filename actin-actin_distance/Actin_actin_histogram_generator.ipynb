```python
"""
Inter-actin Distances of Tip and Shaft Region of Pls-/- Mouse Utricle. Format adapted from Shin et el. 
Three locations of H10_16 were taken and the inter-actin distances were measured. 
The locations were plane 400, plane 800, plane 1190.

Middle region is colored red and the remaining two regions are gray.
Histogram is taken from the middle region.
"""
#set figure size
plt.figure(figsize=(10,8))

location_1 = pd.read_csv('TipShaft01.csv', header=None)
location_2 = pd.read_csv('TipShaft02.csv', header=None)
location_3 = pd.read_csv('TipShaft03.csv', header=None)

# mode_1 = stats.mode(location_1.values, axis=None)[0][0]/10
# mode_2 = stats.mode(location_2.values, axis=None)[0][0]/10
# mode_3 = stats.mode(location_3.values, axis=None)[0][0]/10

nparam_density = stats.kde.gaussian_kde(location_1.values.ravel())
space1 = np.linspace(0, 162, 938)
nparam_density = nparam_density(space1)
peak1 = np.round(space1[np.argsort(nparam_density)[-1]]/10, 2)

nparam_density = stats.kde.gaussian_kde(location_2.values.ravel())
space2 = np.linspace(0, 158.9, 943)
nparam_density = nparam_density(space2)
peak2 = np.round(space2[np.argsort(nparam_density)[-1]]/10, 2)

nparam_density = stats.kde.gaussian_kde(location_3.values.ravel())
space3 = np.linspace(0, 186.9, 922)
nparam_density = nparam_density(space3)
peak3 = np.round(space3[np.argsort(nparam_density)[-1]]/10, 2)

sns.distplot(a=location_1.values/10, hist=False, label=' ', color='#daa520', kde_kws={'linewidth':3})
sns.distplot(a=location_2.values/10, hist=False, label=' ', color='#fa8072', kde_kws={'linewidth':3})
sns.distplot(a=location_3.values/10, hist=False, label=' ', color='#6495ed', kde_kws={'linewidth':3})
# sns.distplot(a=location_2.values/10, hist_kws={'rwidth':0.5, 'color':'gray'}, kde=False, norm_hist=True)

# plt.axvline(x = mode_1, color='#daa520')
# plt.axvline(x = mode_2, color='#fa8072')
# plt.axvline(x = mode_3, color='#6495ed')

plt.legend(fontsize=20)
plt.xlabel("Actin Actin Distances (nm)", fontsize=20)
plt.title("Inter-actin Distances of Tip and Shaft Region of Pls-/- Mouse Utricle", fontsize=25)
plt.text(8.85, 0.355, '{}'.format(peak1), fontsize=15, color='#daa520')
plt.text(8.85, 0.335, '{}'.format(peak2), fontsize=15, color='#fa8072')
plt.text(8.85, 0.315, '{}'.format(peak3), fontsize=15, color='#6495ed')
```

```python
"""
Inter-actin distances of Taper Region of Pls-/- Mouse Utricle. Format adapted from Shin et el.
Three locations of H10_9 were taken and the inter-actin distances were measured. 

Middle region is colored red and the remaining two regions are gray.
Histogram is taken from the middle region.
"""

#set figure size
plt.figure(figsize=(10,8))

tp_location_1 = pd.read_csv('Taper01.csv', header=None)
tp_location_2 = pd.read_csv('Taper02.csv', header=None)
tp_location_3 = pd.read_csv('Taper03.csv', header=None)

# tp_mode_1 = stats.mode(tp_location_1.values, axis=None)[0][0]/10
# tp_mode_2 = stats.mode(tp_location_2.values, axis=None)[0][0]/10
# tp_mode_3 = stats.mode(tp_location_3.values, axis=None)[0][0]/10

nparam_density = stats.kde.gaussian_kde(tp_location_1.values.ravel())
tp_space1 = np.linspace(0, 329.8, 655)
nparam_density = nparam_density(tp_space1)
tp_peak1 = np.round(tp_space1[np.argsort(nparam_density)[-1]]/10, 2)

nparam_density = stats.kde.gaussian_kde(tp_location_2.values.ravel())
tp_space2 = np.linspace(0, 382.6, 256)
nparam_density = nparam_density(tp_space2)
tp_peak2 = np.round(tp_space2[np.argsort(nparam_density)[-1]]/10, 2)

nparam_density = stats.kde.gaussian_kde(tp_location_3.values.ravel())
tp_space3 = np.linspace(0, 421.9, 268)
nparam_density = nparam_density(tp_space3)
tp_peak3 = np.round(tp_space3[np.argsort(nparam_density)[-1]]/10, 2)

sns.distplot(a=tp_location_1.values/10, hist=False, label=' ', color='#daa520', kde_kws={'linewidth':3})
sns.distplot(a=tp_location_2.values/10, hist=False, label=' ', color='#fa8072', kde_kws={'linewidth':3})
sns.distplot(a=tp_location_3.values/10, hist=False, label=' ', color='#6495ed', kde_kws={'linewidth':3})
# sns.distplot(a=tp_location_2.values/10, hist_kws={'rwidth':0.5, 'color':'gray'}, kde=False, norm_hist=True)

# plt.axvline(x = tp_mode_1/10, color='#daa520')
# plt.axvline(x = tp_mode_2/10, color='#fa8072')
# plt.axvline(x = tp_mode_3/10, color='#6495ed')

plt.legend(fontsize=20)
plt.xlabel("Actin Actin Distances (nm)", fontsize=20)
plt.title("Inter-actin Distances of Taper Region of Pls-/- Mouse Utricle", fontsize=25)
plt.text(16.2, 0.1095, '{}'.format(tp_peak1), fontsize=15, color='#daa520')
plt.text(16.2, 0.1015, '{}'.format(tp_peak2), fontsize=15, color='#fa8072')
plt.text(16.2, 0.0935, '{}'.format(tp_peak3), fontsize=15, color='#6495ed')
```
